# Titanic Analysis (Practice for Data Visualization)  
Data source: https://www.kaggle.com/competitions/titanic/data

<u> 1. Import packages and read data. </u>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

train_data = pd.read_csv('titanic_data//train.csv')
```

<u> 2. Basic data visualization. </u>

```python
train_data.head(5)
```

```python
# Countplot for counting the number of people according to the categorical column "Survived"
sns.countplot(x='Survived', data=train_data)
plt.show()
```

```python
sns.countplot(x='Survived', data=train_data, hue='Sex')
plt.show()
```

```python
# Boxplot
sns.boxplot(x='Pclass', y='Age', data=train_data)
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Age', data=train_data, hue='Survived')
plt.show()
```

```python
# Violinplot
sns.violinplot(x='Pclass', y='Age', data=train_data)
plt.show()
```

```python
# Histogram to show the distribution of "Age"
sns.histplot(x='Age', data=train_data)
plt.show()
```

```python
# Pie chart
plt.pie(train_data['Pclass'].value_counts(), labels=train_data['Pclass'].value_counts().index, autopct='%1.2f%%')
plt.show()
```
