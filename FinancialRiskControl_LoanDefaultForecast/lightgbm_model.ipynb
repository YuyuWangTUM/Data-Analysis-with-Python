```python
num_cols = [col for col in train_data.select_dtypes(exclude='object')]
cat_cols = [col for col in train_data.select_dtypes(include='object')]
```

```python
train_data[num_cols] = train_data[num_cols].fillna(train_data[num_cols].median())
```

```python
train_data['employmentLength'] = train_data['employmentLength'].fillna(-1)
```

```python
train_data.info()
```

```python
#!pip install lightgbm
import lightgbm as lgb
from sklearn.model_selection import train_test_split
```

```python
X = train_data.drop('isDefault', axis=1)
X.drop(cat_cols, axis=1, inplace=True)
y = train_data['isDefault']

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
```

```python
train_mat = lgb.Dataset(X_train, label=y_train)
val_mat = lgb.Dataset(X_val, label=y_val)
```

```python
param = {'num_leaves': 31,
        'objective': 'binary',
        'metric': 'auc',
        'boosting_type': 'gbdt',
        'learning_rate': 0.1,
        'min_child_weight': 0.001,
        'max_depth': 5,
        'feature_fraction': 0.8,
        'bagging_fraction': 0.8,
        'seed': 2022,
        'nthread': 8,
        #'silent': True,
        'verbose': -1
        }

lgb_model = lgb.train(param, train_set=train_mat, valid_sets=val_mat, num_boost_round=20000, 
                     verbose_eval=1000, early_stopping_rounds=200)
```

```python
val_pred = lgb_model.predict(X_val, num_iteration=lgb_model.best_iteration)
```

```python
from sklearn.metrics import roc_auc_score
print(roc_auc_score(y_val, val_pred))
```
